```python
from snapi_py_client.snapi_bridge import StocknoteAPIPythonBridge
import json 

import pandas as pd
import numpy as np
from datetime import datetime, date, timedelta, time
import time as Time
import re

def split_alpha_numeric(s):
    # Using regular expression to split alpha and numeric parts
    alpha_part = re.search('[A-Za-z]+', s).group()
    num_part = re.search('[0-9]+', s).group()
    return alpha_part

while datetime.now().time() < time(15, 35, 30):
    Time.sleep(1)

samco=StocknoteAPIPythonBridge()
login=json.loads(samco.login(body={"userId":'DL31725','password':'Bharat@9998','yob':'1998'}))
samco.set_session_token(sessionToken=login['sessionToken'])

derivative = json.loads(samco.search_equity_derivative(search_symbol_name="NIFTY",exchange=samco.EXCHANGE_NFO))['searchResults']
DF = pd.DataFrame(derivative)

OPTIDX = DF[DF['instrument']=='OPTIDX']

# Applying function to DataFrame
OPTIDX['INDEX'] = OPTIDX['tradingSymbol'].apply(split_alpha_numeric)

index_list = ['NIFTY']#,'BANKNIFTY','FINNIFTY']
data_not_found = []

for index in index_list:
    
    symbols = OPTIDX[OPTIDX['INDEX']==index]['tradingSymbol'].to_list()

    main_data = pd.DataFrame()

    start_date = str(date.today())+' 09:15:00'
    end_date = str(date.today())+' 15:30:00'
    
    for i in symbols:
        print(i)

        try:
            data = pd.DataFrame(json.loads(samco.get_intraday_candle_data(symbol_name=i,exchange=samco.EXCHANGE_NFO, from_date=start_date, to_date=end_date))['intradayCandleData'])
            data['symbol'] = i   #'NIFTY05JAN2318300CE'
            main_data = pd.concat([main_data,data],axis=0)
        except:
            data_not_found.append(i)
            continue
            
    for trying in range(5):
        for j in data_not_found:
            print(j,'trying---->',trying)

            try:
                data = pd.DataFrame(json.loads(samco.get_intraday_candle_data(symbol_name=i,exchange=samco.EXCHANGE_NFO, from_date=start_date, to_date=end_date))['intradayCandleData'])
                data['symbol'] = i   #'NIFTY05JAN2318300CE'
                main_data = pd.concat([main_data,data],axis=0)
                data_not_found.remove(i)
            except:
                continue


    main_data['dateTime'] = pd.to_datetime(main_data['dateTime'])
    main_data.sort_values(by='dateTime',inplace=True)

    path =  fr'E:/option_data/INDEX/{index}/2024/FEB/data_'
    dates = path+str(date.today())+'.csv'

    main_data.to_csv(dates,index=False)
```

### Move file one folder to another 

```python
import shutil

index_list = ['NIFTY','BANKNIFTY','FINNIFTY']

dates = ['2024-02-01','2024-02-02','2024-02-05']

for instument in index_list:
    path = fr'E:\option_data\INDEX\{instument}\2024\JAN\data_'
    
    for D in dates:
        source_path = path+str(D)+'.csv'
    
        destination_path = fr'E:/option_data/INDEX/{instument}/2024/FEB/'

        # Use shutil.move to move the file
        shutil.move(source_path, destination_path)
```
